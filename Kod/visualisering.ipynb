```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\daaro\OneDrive - Karlskoga Kommun\Studier\YH\AI och IoT\Kunskapskontroll 1\AI Kunskapskontroll 2\iot_data.csv")
```

```python
df["timestamp"] = pd.to_datetime(df["timestamp"], unit="s")
df = df.sort_values("timestamp")

plt.figure(figsize=(12, 5))
plt.plot(df["timestamp"], df["temperature"], marker="o", linestyle="-")
plt.title("Temperatur över tid")
plt.xlabel("Tid")
plt.ylabel("Temperatur")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
